```python
from openai import OpenAI

import os
from dotenv import load_dotenv
load_dotenv()

client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY")
)

prompt = "Write a 5000-word epic fantasy story about the rise and fall of an ancient empire. Include detailed character dialogues, historical background, cultural traditions, battle descriptions, and a philosophical ending. Ensure that the text is continuous and does not stop prematurely."

for i in range(5):
    response = client.responses.create(
        model="o1-pro",
        input=prompt,
        max_output_tokens=8000
    )
    print(f"Run {i} done")
    print(response)
```
